```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
from pandas import read_excel
import bqplot
```

```python
# find your sheet name at the bottom left of your excel file and assign 
# it to sheet_name
my_sheet = 'Hygiene'

# name of your excel file
file_name = '/Users/beenabalakrishna/Desktop/UIUC /SEM 2 /PR /Assignments/Final project /Dataset/UNICEF2017.xlsx' 
who_unicef = pd.read_excel(file_name, sheet_name= my_sheet)

who_unicef
```

As we can observe, there are 39 columns in this dataset representing the 'Hygiene' quality maintained in 76 countries further categorized into Urban, Rural, Hospital, Non-Hospital, Government and Non-Government areas.

```python
wb_data = pd.read_csv("/Users/beenabalakrishna/Desktop/UIUC /SEM 2 /PR /Assignments/Final project /Dataset/world_bank_2017/Income_classification.csv")
wb_data['IncomeGroup'] = wb_data['IncomeGroup'].replace(np.nan, 0)
wb_data
```

```python
merged = pd.merge(who_unicef, wb_data)
merged
```

```python
merg = merged.iloc[:,0:3]
merg
```

```python
x = merg.CountryName
y = merg.Population(thousands)/1000
```

After merging the two datasets, it can noticed that the new dataframe consists of 64 rows and 44 columns. 

```python
plt.figure(figsize = (15,20))
plt.bar(x,y,aplha = 0.5, colour = 'purple')
plt.xticks(rotation=90)
plt.xlabel("Country")
plt.ylabel("Population in thousands")
plt.title("Population of the countires")
plt.show()
```
