```python
wb_data = pd.read_csv("/Users/beenabalakrishna/Desktop/UIUC /SEM 2 /PR /Assignments/Final project /Dataset/world_bank_2017/Income_classification.csv")
wb_data['IncomeGroup'] = wb_data['IncomeGroup'].replace(np.nan, 0)
wb_data
```

```python
merged = pd.merge(who_unicef, wb_data)
merged
```

```python
merged["CountryEconomicStatus"] = np.where(np.logical_or(merged['IncomeGroup']=='High income', merged['IncomeGroup']=='Upper middle income'), 'Developed', 'Developing')
```

```python
merged
```




**Note:** The column "IncomeGroup" is merged into two groups under the "CountryEconomicStatus" column namely; Developed and developing. The classification is done based on the GNI (Gross National Income) of the countries and the reason why I have chosen GNI as a parameter is because 

`GNI = GDP + (wages + salary + property income) of a country.`

```python
merg = merged.loc[:,["CountryName", "Population(thousands)", "CountryEconomicStatus"]]

merg 
```

```python
x = merg.CountryName
y = merg['Population(thousands)']
```

```python
plt.figure(figsize=(25,25))
ax = plt.gca()
ax.set_facecolor("xkcd:white")
plt.bar(x,y,alpha = 0.8, color='xkcd:purple')
ax.set_yscale('log')
plt.xticks(rotation=90)
plt.xlabel("Country Name", fontsize = 20)
plt.ylabel("Population (log scale)", fontsize = 20)
plt.title("Population of the countires", fontsize = 30)
```

**The above bar plot shows the population (in logarithmic scale) of 63 countries.** 

```python
def status ( x ):
    developed = pd.DataFrame()
    developing = pd.DataFrame()
    for i in range(x.shape[0]):
        if (x.iloc[i]['CountryEconomicStatus'] == "Developed"): 
            # define developed dataframe here
            data = pd.DataFrame(dict(merg.iloc[i]), index=[0])
            developed = developed.append(data)
        else: 
            # define developing dataframe here
            data = pd.DataFrame(dict(merg.iloc[i]), index = [0])
            developing = developing.append(data)
    return (developed, developing)
  
# Driver code 
developed, developing = status(merg) 
```
